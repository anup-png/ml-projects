```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('tip',axis=1),df['tip'],test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

preProcessor =ColumnTransformer(
    transformers=[
        ('enc',OneHotEncoder(),['sex','smoker','day','time'])
    ],
    remainder='passthrough'
)

df_new =preProcessor.fit_transform(x_train)
df_test_new =preProcessor.fit_transform(x_test)
features_name =preProcessor.get_feature_names_out()

tranformed_df =pd.DataFrame(df_new,columns=features_name)
tranformed_test_df =pd.DataFrame(df_test_new,columns=features_name)
```

```python
tranformed_df
```

```python
tranformed_test_df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
model = lr.fit(tranformed_df,y_train)
result = model.predict(tranformed_test_df)

acc = r2_score(y_test,result
)
acc
```
